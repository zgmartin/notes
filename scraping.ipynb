```python
fin = pandas.read_csv('/Users/gold/Downloads/russell_2000_finances.csv')
fin = fin.drop(columns=['Unnamed: 0'])
```

```python
fin
```

```python
form = {'symbol': [], 'sector': [], 'industry': [], 'employees': []}
```

```python
for item in fin.symbol.unique():
    print(item)
    symbol = item
    url = 'https://finance.yahoo.com/quote/{}/profile?p{}'.format(symbol,symbol)
    html = requests.get(url)
    soup = bs4.BeautifulSoup(html.text)
    
    tags = soup.find_all("span", {"class": "Fw(600)"})
    columns = ['symbol','sector', 'industry', 'employees']
    data = []
    
    data.append(symbol)
    for i,tag in enumerate(tags):
        try:
            content = tag.contents[0]
            if i is 2: content = content.contents[0]
            data.append(content)
        except:
            data.append(None)
    if len(data)<3: data = [symbol,None,None,None]
    
    data = list(zip(columns, data))
    
    for item in data:
        form[item[0]].append(item[1]) 
```

```python
info = pandas.DataFrame(form)
info
```

```python
fins = fin.merge(info, on='symbol')
```

```python
data = fins
```

```python
for i,item in enumerate(data.employees):
    if item is not None:
        data.at[i,'employees'] = item.replace(',','')
        
data.fillna(value='NULL', inplace=True)
data.replace(to_replace='-',value='NULL', inplace=True)
```
